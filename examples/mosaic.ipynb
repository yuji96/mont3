```python
import numpy as np
import oekaki
```

```python
fig = oekaki.figure(constrained_layout=True, level="ignore")

fig["bar"].bar(["a", "b", "c"], [5, 7, 9])
fig["image"].imshow([[1, 2], [2, 1]])
fig["line"].sns.lineplot(data=np.random.rand(100))
fig["hist"].sns.histplot(data=np.random.rand(100))
```

```python
fig["image|line"].set(ylim=[0, 1])
# fig[".*"].set(facecolor="azure")
fig.show("""bar  | image
            line | line """)
```
