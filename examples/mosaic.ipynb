```python
import numpy as np
import oekaki
```

```python
fig = oekaki.figure(constrained_layout=True, level="ignore")
fig["bar"].bar(["a", "b", "c"], [5, 7, 9])
fig["plot"].plot([1, 2, 3])
fig["hist"].hist(np.random.rand(100))
fig["image"].imshow([[1, 2], [2, 1]])
```

```python
fig.show("""bar  | plot
            hist | hist """)
```

```python
fig.show("""bar  | plot
            hist | image """)
```
